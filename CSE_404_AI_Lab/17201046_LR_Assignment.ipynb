```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np

db = load_diabetes()
```

```python
X = db.data
y = db.target
```

```python
X.shape
```

```python
df = pd.DataFrame(X)
df = pd.concat((df,pd.DataFrame(y,columns=['y'])),axis=1)
df.head()
```

```python
df.columns=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','y']
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
def LR(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,y,t0,t1,t2,t3,t4,t5,t6,t7,t8,t9,t10):
    hat =[]
    ls =[]
    for a,b,c,d,e,f,g,h,i,j in zip(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10):
        hat.append(t0 + a*t1 + b*t2 + c*t3 + d*t4 + e*t5 + f*t6 + g*t7 + h*t8 + i*t9 + j*t10)
        
    for i in range(len(y)):
        ls.append((y[i]-hat[i])**2)

    return hat,ls
```

```python
x1,x2,x3,x4,x5,x6,x7,x8,x9,x10 = list(df.x1),list(df.x2),list(df.x3),list(df.x4),list(df.x5),list(df.x6),list(df.x7),list(df.x8),list(df.x9),list(df.x10)
```

```python
t0,t1,t2,t3,t4,t5,t6,t7,t8,t9,t10 = 1.308343,3.634285,1.308343,8.04534,1.28165,8.8353,1.3270,4.57464,3.7773,3.830,3.4128
```

```python
for i in range(10):
    hat,dist = LR(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,y,t0,t1,t2,t3,t4,t5,t6,t7,t8,t9,t10)
    df['prediction'] = hat
    df['distance_sqrd'] = dist
    
    m = df.shape[0]
    cost = (df['distance_sqrd'].sum())/(2*m)
    df['dist'] = np.sqrt(df['distance_sqrd'])
    print('For step : ',i)
    print('Cost is : ',cost)
    

    
    step_size = 0.001

    t0 = t0 - (step_size/m)*(df['dist'].sum())
    t1 = t1 - (step_size/m)*((df['dist']*x1).sum())
    t2 = t2 - (step_size/m)*((df['dist']*x2).sum())
    t3 = t3 - (step_size/m)*((df['dist']*x3).sum())
    t4 = t4 - (step_size/m)*((df['dist']*x4).sum())
    t5 = t5 - (step_size/m)*((df['dist']*x5).sum())
    t6 = t6 - (step_size/m)*((df['dist']*x6).sum())
    t7 = t7 - (step_size/m)*((df['dist']*x7).sum())
    t8 = t8 - (step_size/m)*((df['dist']*x8).sum())
    t9 = t9 - (step_size/m)*((df['dist']*x9).sum())
    t10 = t10 - (step_size/m)*((df['dist']*x10).sum())

    print('After gradient decent optimization new thetas value : ',t0,t1,t2,t3,t4,t5,t6,t7,t8,t9,t10)
    print('\n\n')
```
